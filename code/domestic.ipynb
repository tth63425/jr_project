```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('../data/clean/Book1_semi.xlsx')
```

```python
df
```

```python
# stay_all

cpc_list = []

for i in range(423):
    consumption = df.iloc[i, 3]
    population = df.iloc[i, 11]
    
    if consumption != 0 and population != 0:
        cpc = consumption / population
    else:
        cpc = 0
        
    cpc_list.append(cpc)
```

```python
df["cpc_stay_all"] = cpc_list
```
