```python
import pandas as pd
import numpy as np
```

```python
# by prefecture part 1

df_pref_part1 = []

for i in range(16):
    sheet_number = i + 1
    
    if sheet_number % 2 == 0:
        # prefecture
        counter = 0
        sheet = pd.read_excel('../data/191011_191221kameiten_touroku_report.xlsx', sheet_name=i)
        df_pref_part1.append(sheet)
        counter += 1
```

```python
# by prefecture part 2

df_pref_part2 = []

for i in range(22):
    sheet_number = i + 1
    
    if sheet_number % 2 == 0:
        # prefecture
        counter = 0
        sheet = pd.read_excel('../data/200111_200611kameiten_touroku_report.xlsx', sheet_name=i)
        df_pref_part2.append(sheet)
        counter += 1
```

```python
df_pref_part1[1]
```

```python
# 加盟店数
kameiten_all = []

for i in range(8):
    list_kamei = list(df_pref_part1[i].iloc[3:1791,3])
    kameiten_all.append(list_kamei)
    
for i in range(11):
    list_kamei2 = list(df_pref_part2[i].iloc[3:1791,3])
    kameiten_all.append(list_kamei2)
```

```python
# 固定店舗

koteitenn_all = []

for i in range(8):
    list_kotei = list(df_pref_part1[i].iloc[3:1791,4])
    koteitenn_all.append(list_kotei)
    
for i in range(11):
    list_kotei2 = list(df_pref_part2[i].iloc[3:1791,4])
    koteitenn_all.append(list_kotei2)
```

```python
# 飲食業

insyoku_all = []

for i in range(8):
    list_insyoku = list(df_pref_part1[i].iloc[3:1791,5])
    insyoku_all.append(list_insyoku)
    
for i in range(11):
    list_insyoku2 = list(df_pref_part2[i].iloc[3:1791,5])
    insyoku_all.append(list_insyoku2)
```

```python
# その他サービス業

service_all = []

for i in range(8):
    list_service = list(df_pref_part1[i].iloc[3:1791,6])
    service_all.append(list_service)
    
for i in range(11):
    list_service2 = list(df_pref_part2[i].iloc[3:1791,6])
    service_all.append(list_service2)
```

```python
# 通信

tele_all = []

for i in range(8):
    list_tele = list(df_pref_part1[i].iloc[3:1791,7])
    tele_all.append(list_tele)
    
for i in range(11):
    list_tele2 = list(df_pref_part2[i].iloc[3:1791,7])
    tele_all.append(list_tele2)
```

```python
# 移動販売

idou_all = []

for i in range(8):
    list_idou = list(df_pref_part1[i].iloc[3:1791,8])
    idou_all.append(list_idou)
    
for i in range(11):
    list_idou2 = list(df_pref_part2[i].iloc[3:1791,8])
    idou_all.append(list_idou2)
```

```python
prefs_1 = list(df_pref_part1[2].iloc[3:1791,1])
prefs_2 = list(df_pref_part1[2].iloc[3:1791,2])
prefs = []

for i in range(1788):
    if not type(prefs_2[i]) == float:
        str_pref = str(prefs_1[i]) + str(prefs_2[i])
        prefs.append(str_pref)
    else:
        prefs.append(prefs_1[i])
```

```python
times = ["1011","1021","1101","1111","1121","1201","1211","1221","0111","0121","0201","0211","0221","0301","0311","0321","0401","0601","0611"]
```

```python
kameitan_array = np.array(kameiten_all)
kameitan_array = kameitan_array.T
df_kameiten = pd.DataFrame(kameitan_array, columns=times, index=prefs)
df_kameiten.to_csv("../csv/sikutyouson/kameiten.csv")
```

```python
koteitenn_array = np.array(koteitenn_all)
koteitenn_array = koteitenn_array.T
df_koteitenn = pd.DataFrame(koteitenn_array, columns=times, index=prefs)
df_koteitenn.to_csv("../csv/sikutyouson/koteitenn.csv")
```

```python
insyoku_array = np.array(insyoku_all)
insyoku_array = insyoku_array.T
df_insyoku = pd.DataFrame(insyoku_array, columns=times, index=prefs)
df_insyoku.to_csv("../csv/sikutyouson/insyoku.csv")
```

```python
service_array = np.array(service_all)
service_array = service_array.T
df_service = pd.DataFrame(service_array, columns=times, index=prefs)
df_service.to_csv("../csv/sikutyouson/service.csv")
```

```python
tele_array = np.array(tele_all)
tele_array = tele_array.T
df_tele = pd.DataFrame(tele_array, columns=times, index=prefs)
df_tele.to_csv("../csv/sikutyouson/tele.csv")
```

```python
idou_array = np.array(idou_all)
idou_array = idou_array.T
df_idou = pd.DataFrame(idou_array, columns=times, index=prefs)
df_idou.to_csv("../csv/sikutyouson/idou.csv")
```
