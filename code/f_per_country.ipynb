```python
# loop for country
counter = 0
for i in range(7, 49, 2):
    country = list_countries[counter]
    dict_row = {"year": 2018, "term": 1, "country": country}
    
    consumption = df.iloc[5, i]
    dict_tmp = {"c_all":consumption}
    dict_row.update(dict_tmp)
    
    # loop for type of consunption
    for j in range(len(list_columns1)):
        type_consumption = list_columns1[j]
        consumption = df.iloc[41+j, i]
        dict_tmp = {"c_" + type_consumption : consumption}
        dict_row.update(dict_tmp)
        
    for k in range(len(list_column2)):
        type_consumption = list_column2[k]
        consumption = df.iloc[107+k, i]
        dict_tmp = {"c_" + type_consumption : consumption}
        dict_row.update(dict_tmp)
        
    id_column = i - 1
    answer = df.iloc[5, id_column]
    dict_tmp = {"a_all":answer}
    dict_row.update(dict_tmp)
    
    # loop for type of consunption
    for j in range(len(list_columns1)):
        type_consumption = list_columns1[j]
        consumption = df.iloc[41+j, id_column]
        dict_tmp = {"a_" + type_consumption : consumption}
        dict_row.update(dict_tmp)
        
    for k in range(len(list_column2)):
        type_consumption = list_column2[k]
        consumption = df.iloc[107+k, id_column]
        dict_tmp = {"a_" + type_consumption : consumption}
        dict_row.update(dict_tmp)
    
    counter = counter + 1
    dict_all.append(dict_row)
```

```python
tmp_list = list(np.zeros(76, dtype="int8"))
```

```python
test_keys = list_columns
test_values = tmp_list
```

```python
res = {} 
for key in test_keys: 
    for value in test_values: 
        res[key] = value 
        test_values.remove(value) 
        break 
```

```python
dict_all = []

for i in range(df.shape[0]):
    
    for j in range(4, 80):
        dict_tmp = {
            "year" : df.iloc[i,1],
            "term" : df.iloc[i,2],
            "country" : df.iloc[i, 3],
        }
        
        consumption = df.iloc[i, j]
        answer = df.iloc[i, j + 76]
        
        dict2 = {
            "consumption" : consumption,
            "answer" : answer
        }
        
        type_element = list_columns[j - 4]
        
        dict_tmp.update(dict2)
        dict_tmp.update(res)
        dict_tmp[type_element] = 1
        
        dict_all.append(dict_tmp)
```
